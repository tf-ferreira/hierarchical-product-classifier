# 02 · Análise de erros

Um treino sem análise de erros é meio treino. Aqui interpretamos (não apenas
plotamos) as confusões dos dois runs. Rode APÓS `hierclf.evaluate` em cada run.

```python
import glob, json
import pandas as pd

runs = sorted(glob.glob("../experiments/*-*/evaluation.json"))
evals = [json.load(open(p)) for p in runs]
pd.DataFrame([{k: v for k, v in e.items() if not isinstance(v, list)} for e in evals])
```
