Objetivo: entender a estrutura hierárquica, o desbalanceamento e a qualidade dos dados
ANTES de treinar. Toda a lógica reutilizável vive em `hierclf`; o notebook só narra.

```python
from hierclf.data import download_dataset, load_and_validate

data_root = download_dataset(version="small")
df, taxonomy, report = load_and_validate(data_root, min_samples_per_class=20)
print(report.resumo())
report.conflitos_taxonomia
```

## Distribuição das classes por nível
A cauda longa do `articleType` é o fato central do dataset: justifica o split
estratificado e o corte de classes raras (ambos em `hierclf.data`).

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
for ax, level in zip(axes, ["masterCategory", "subCategory", "articleType"]):
    df[level].value_counts().plot(kind="bar", ax=ax, title=f"{level} ({df[level].nunique()} classes)")
    ax.set_xticklabels([] if level == "articleType" else ax.get_xticklabels(), rotation=90, fontsize=7)
plt.tight_layout()
plt.savefig("../reports/figures/distribuicao_classes.png", dpi=150, bbox_inches="tight")
```

## Amostras por categoria principal

```python
from PIL import Image
sample = df.groupby("masterCategory").head(3)
fig, axes = plt.subplots(len(sample)//3, 3, figsize=(9, 3*len(sample)//3))
for ax, (_, row) in zip(axes.flat, sample.iterrows()):
    ax.imshow(Image.open(row.image_path)); ax.axis("off")
    ax.set_title(f'{row.masterCategory} / {row.articleType}', fontsize=8)
plt.tight_layout()
```
